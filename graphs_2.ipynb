```python
import matplotlib.pyplot as plt
import seaborn as sns
from jupyterthemes import jtplot
jtplot.style(theme='onedork', context='notebook', ticks=True, grid=False)
```

```python
y = [6, 8, 3, 1, 9]
x1 = [1, 2, 3, 4, 5]
x = ['seg', 'ter', 'qua', 'qui', 'sex']
```
